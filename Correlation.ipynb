```python
corr_matrix_practice = np.array([[1,0.22,0.65],
                                 [0.22,1,-0.92],
                                [0.65,-0.92,1]])
```

```python
for row in range(len(corr_matrix_practice)):
    for col in range(row):
        print(corr_matrix_practice[row][col])
```

```python
for row in range(len(corr_matrix_practice)):
    for col in range(row):
        if corr_matrix_practice[row][col] > 0.5:
            print(corr_matrix_practice[row][col])
```

```python
for row in range(len(corr_matrix_practice)):
    for col in range(row):
        if abs(corr_matrix_practice[row][col]) > 0.5:
            print(corr_matrix_practice[row][col])
            
```

```python
for row in range(len(corr_matrix)):
    for col in range(row):
        if abs(corr_matrix.iloc[row][col]) > 0.5:
            print(corr_matrix.iloc[row][col])
            print(corr_matrix.columns[row])
            
            
```

```python
corr_names = set()
for row in range(len(corr_matrix)):
    for col in range(row):
        if abs(corr_matrix.iloc[row][col]) > 0.5:
            corr_names.add(corr_matrix.columns[row])
            
            
```

```python
corr_names
```

```python
df
```

```python
df.drop(corr_names,axis = 1, inplace=True)
```
